<i>Условие:</i>
<br>Саня и Ваня подбрасывают монету. Монета симметрична, и шанс выпадения орла и решки равны. Саня подбрасывает её $n$ раз, а Ваня $n+1$ раз. Какова вероятность того, что у Вани выпадет больше орлов, чем у Сани.

<i>Решение:</i>

Пусть Саня и Ваня подбросили монету $n$ раз. Тогда имеем 3 исхода: 
<br>1) У Сани орел выпал больше раз, чем у Вани
<br>2) У Вани орел выпал больше раз, чем у Сани
<br>3) И У Вани, и у Сани орел выпал одинаковое количество раз.
<br>Очевидно, что случаи 1 и 2 равновероятны. Тогда вероятность 3-его случая равна $1 - 2p$ (где p - вероятность 1-ого или 2-ого случая)
<br>Подбрасываем монету еще раз. Если у Вани стало больше орлов, чем у Сани, то на предыдущем "ходе" у него итак было преимущество: орел выпал больше раз, чем у Сани, либо у них было по одинаковому числу выпадения орлов. Но эти события несовместны. Вероятность первого события $= p$, а вероятность второго события $= \dfrac{1-2p}{2}$ (т.к. монета симметрична и шанс выпадения орла $= 0.5$
Тогда по правилу сложения (т.к. события несовместны) имеем:
$$ p + \dfrac{1-2p}{2} = \dfrac{2p + 1 - 2p}{2} = 0.5 $$

<i>Условие:</i>
<br>Сколько способов пройти из $(0, 0, 0)$ в $(n, 3n, 5n)$, если можно делать шаги на $+1$ по любой из осей?

<i>Решение:</i>

Так как можно делать шаги лишь на $+1$ по любой из осей, то в точку $(n, 3n, 5n)$ можно прийти за $9n$ шагов. Из этих $9n$ выберем $n$: шаги по оси $Ox$:
$${9n \choose n} = \frac{(9n)!}{n!  \cdot (8n)!}$$
Теперь из оставшихся $8n$ шагов нужно выбрать $3n$ по оси $Oy$:
$${8n \choose 3n} = \frac{(8n)!}{3n!  \cdot (5n)!}$$
Остались лишь $5n$ шагов по оси $Oz$. Очевидно, что количество способов их сделать $= 1$. Пользуясь комбинаторным правилом умножения имеем:
$$1 \cdot \frac{(9n)!}{n!  \cdot (8n)!} \cdot \frac{(8n)!}{3n!  \cdot (5n)!} = \dfrac{(9n)!}{n! \cdot (3n)! \cdot (5n)!} $$

<i>Условие:</i>
<br>Команда Джека Воробья в количестве 6 пиратов (считая Джека) закопала добытые однажды монеты на острове. Спустя месяц Джек вернулся на остров, разделил монеты на 6 частей. Вышло не поровну: осталась одна монета сверху. Джек взял себе одну кучку и одну монету и закопал остаток. Абсолютно такая же ситуация возникала и у каждого из остальных пяти пиратов, и каждый не подозревал о действиях другого. Через пару лет Дэйви Джонс нашел спрятанный клад. Какое минимальное количество монет его ждало?

<i>Решение:</i>

Пусть изначально было $n$ монет. Спустя месяц, после Джека, их осталось $n - 1 - \frac{n-1}{6} = \frac{5}{6}(n-1)$
<br>Несложно заметить, что подобный паттерн повторяется еще $5$ раз. Сначала из кучи вычитается $1$, а потом умножается на $\frac{5}{6}$. Тогда пусть $k$ – оставшееся кол-во монет, доставшиеся Дэйви Джонсу. Тогда:
$$k = (\frac{5}{6})^{6} \cdot n - \sum_{i=1}{6} (\frac{5}{6})^{i}$$
Отсюда имеем:
$$n = \frac{k + \sum_{i=1}{6} (\frac{5}{6})^{i}}{(\frac{5}{6})^{6}}$$

Тогда, составив программу, находим $n$ и $k$:

```python
# k - клад
# n - изначальное кол-во монет
# k = (5/6)^6 * n - sum( (5/6)^i, i=1..6)
# a = (5/6)^6 = 15625/46656
# b = sum( (5/6)^i, i=1..6) = 155155/46656
# n = (k + b)/a: k, n: целые

import math


k = 1
a = 15625/46656
b = 155155/46656
n = (k+b)/a
print('start: {} {}'.format(k, n))

while math.modf(n)[0] != 0.0:  # пока n - не целое
    k += 1
    n = (k+b)/a
    #print('{} {}'.format(k, n))

print('Найденное число: {}'.format(n))
print('Проверка:-------------')

for i in range(6):
    print('остаток:{}'.format(n % 6))
    n -= 1
    n -= n//6
    print(n)

print('Ответ: {}'.format(k))
```

<i>Решение:</i>

Пусть в компании $n$ человек: $K= \left \{1,2,…,n\right \}$
Следовательно у каждого человека на вечеринке не может быть более чем $n – 1$ знакомых: $L=\left \{1,2,…,n-1\right \}⊂N$-кол-во знакомых
Построим отображение $f: \left \{1,2,…,n\right \}  → \left \{1,2,…,n-1\right \} $ 
Очевидно, что оно не является уж точно инъективным, а значит существуют хотя бы $2$ человека, имеющих равное количество знакомых в компании
Непосредственно к решению также приводит и принцип Дирихле: в роли клеток – множество $L$, а в роли зайцев – $K$. Очевидно, что как минимум в одной клетке будут находиться как минимум $2$ зайца.

<i>Решение:</i>

Рассмотрим множество $M$:
$M=\left \{11,12,…,23\right \}$-написанные Саней числа
<br>Очевидно, что чтобы произведение чисел давало квадрат некоего числа, степени простых делителей полученного произведения должны быть кратны $2$. Отсюда следует, что на доске точно останется $16 = 2^4$ 
<br>Также несложно сделать вывод, что все простые числа(кроме, быть может $11$), точно будут стерты. Пусть $N$ – множество стертых чисел. Постепенно будет пополнять его значениями. Пока лишь $N=\left \{13,17,19,23\right \}$
Рассмотрим оставшиеся элементы $M$ – оставшиеся на доске числа, и запишем их простые делители:
<br>$11$ – простое число
<br>$12 = 2^2 \cdot 3$
<br>$14 = 2 \cdot 7$
<br>$15=3 \cdot 5$
<br>$16=2^4$ 
<br>$18 = 2 \cdot 3^2$
<br>$20 = 2^2 \cdot 5$
<br>$21 = 3 \cdot 7$
<br>$22 = 2 \cdot 11$
<br>Видно, что если мы оставим $11$, то необходимо оставить и $22$. Но тогда нужно как-то «добить» степень двойки до второй, чтобы произведение вышло квадратом некоторого числа. Есть $2$ варианта:

<br>&emsp;б) Оставляем на доске $15$. Тогда неминуемо необходимо оставить и $20$. На доске остались: $16$, $11$, $22$, $14$, $21$, $15$, $20$. Их произведение $= 3 41 510 400$. Это квадрат числа $18 480$. Следующие числа стерлись: $12$, $13$, $17$, $18$, $19$, $23$ ($6$ чисел). Их сумма $= 102 $
<br>2) Оставляем на доске $18$. Тогда уж точно на доске останутся: $16$, $11$, $22$, $18$. Оставим также на доске еще несколько чисел: оставим $12$(т.к. необходимо лишь найти одну $3$, чтобы произведение все еще являлось неким квадратом). Чтобы покрыть появившуюся $3$ имеем $2$ варианта:

То есть уже минимум $6$ стертых чисел. Так как в задаче не сказано находить минимальную сумму минимально-стертых чисел, то можно обойтись найденными результатами выше.
